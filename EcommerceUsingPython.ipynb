```python
# üõí Mini E-commerce Billing System

# Product catalog with prices
products = {
    "Laptop": 50000,
    "Phone": 20000,
    "Headphones": 3000,
    "Mouse": 1500,
    "Keyboard": 2000
}

# Function to generate bill
def generate_bill(cart, discount=0, tax_percentage=0, customer_name="Guest"):
    print("\nüõí Shopping Bill")
    print("-" * 40)
    
    subtotal = sum(products[item] * qty for item, qty in cart.items())
    for item, qty in cart.items():
        print(f"{item} (x{qty}) - ‚Çπ{products[item] * qty:,}")
    
    print("-" * 40)
    print(f"Subtotal: ‚Çπ{subtotal:,}")
    
    # Apply discount
    if discount > 0:
        subtotal -= discount
        print(f"üéâ Discount Applied: ‚Çπ{discount:,}")
    
    # Apply tax percentage
    tax = (tax_percentage / 100) * subtotal if tax_percentage > 0 else 0
    if tax > 0:
        subtotal += tax
        print(f"üí∞ Tax ({tax_percentage}%): ‚Çπ{tax:,}")
    
    print("-" * 40)
    print(f"üí≥ Final Amount: ‚Çπ{subtotal:,}")
    print(f"üë§ Customer: {customer_name}")
    print("-" * 40)
    print("üõçÔ∏è Thank you for shopping with us! Come back soon!\n")

# Main logic
cart = {}

print("üì¢ Welcome to the Mini E-commerce Store!\n")
print("üõí Available Products:")
for item, price in products.items():
    print(f"‚úÖ {item} - ‚Çπ{price:,}")

while True:
    choice = input("\nEnter product name to add/remove (or 'done' to finish shopping): ").title()
    
    if choice == "Done":
        break
    
    if choice in products:
        action = input("Would you like to 'add' or 'remove' this item? ").lower()
        if action == "add":
            try:
                qty = int(input(f"How many {choice}s would you like? "))
                if qty > 0:
                    cart[choice] = cart.get(choice, 0) + qty
                    print(f"üõçÔ∏è {qty} {choice}(s) added to your cart.")
                else:
                    print("‚ö†Ô∏è Please add at least one item.")
            except ValueError:
                print("üö® Oops! That doesn't look like a number. Try again!")
        elif action == "remove":
            if choice in cart:
                del cart[choice]
                print(f"üóëÔ∏è {choice} removed from your cart.")
            else:
                print("‚ùå That item is not in your cart.")
        else:
            print("üö® Invalid action! Please type 'add' or 'remove'.")
    else:
        print("üö® We don't have that product. Please try again.")

# Display cart before proceeding
if cart:
    print("\nüõí Your Shopping Cart:")
    for item, qty in cart.items():
        print(f"- {item} (x{qty})")
else:
    print("\nüòî Your cart is empty. Come back soon!")
    exit()

# Ask for additional details with validation
customer_name = input("\nWhat is your name? (Press enter to skip): ") or "Guest"

def get_valid_number(prompt, allow_float=False):
    while True:
        try:
            value = float(input(prompt) or 0) if allow_float else int(input(prompt) or 0)
            if value >= 0:
                return value
            else:
                print("‚ö†Ô∏è Please enter a non-negative number.")
        except ValueError:
            print("üö® Invalid input! Try entering a number.")

discount = get_valid_number("Enter discount amount (‚Çπ): ")
tax_percentage = get_valid_number("Enter tax percentage (0-100): ", allow_float=True)

# Confirm checkout
done = input("\nReady to checkout? (yes/no): ").strip().lower()
if done in ["yes", "y"]:
    generate_bill(cart, discount, tax_percentage, customer_name)
else:
    print("\nüõí Checkout canceled. See you next time!")
```
